```python
import pickle
import xlrd

year = 2013
index = 2

s_2015 = xlrd.open_workbook("resp.xlsx").sheet_by_index(index)

label_2015 = {}
for row in range(s_2015.nrows):
    name = s_2015.row(row)[0].value.replace("\x1d", "").replace(" ", "").replace("-", "").replace("\n", "")
    location = str(s_2015.row(row)[1].value)
    if s_2015.row(row)[2].value == 1.0:
        try:
            print(label_2015[name + "_" + location])
            print(name + "_" + location)
        except:
            label_2015[name + "_" + location] = 1
    elif s_2015.row(row)[2].value == 0.0:
        label_2015[name + "_" + location] = 0
    else:
        print(name)
        
print(len(label_2015))
print(label_2015)
```
